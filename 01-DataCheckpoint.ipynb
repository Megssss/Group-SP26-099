## Background and Prior Work

In 2025, the gaming industry reached a record-breaking $190 billion in market size. Gaming is widespread across the world, with players spanning multiple countries and continents, demonstrating that video games are a shared but culturally influenced activity. Among all genres, action games consistently rank as one of the most popular worldwide, making them a useful focus for analyzing player preferences and consumer behavior.

Previous research has explored how review scores influence video game sales performance. A 2024 study found that higher user review scores were associated with a decrease in total video game sales, while higher critic review scores showed an exponential increase in sales.<a name="ref1"></a><sup>1</sup> These findings suggest that professional critical reception may have a stronger impact on commercial success than player reception alone, highlighting the importance of critic influence and reputation within the gaming market.

A 2021 study found that action games consistently generate the highest global sales revenues across all genres, while North America leads the world in consumer spending on video games.<a name="ref1"></a><sup>2</sup> These trends reinforce the importance of studying action games specifically, as they dominate the market and attract broad international audiences. 

Most recently, a 2026 study found that certain publishers consistently produce higher-selling games, suggesting that publisher name-recognition and reputation contribute heavily to each title's commercial success.<a name="ref1"></a><sup>3</sup> However, the study concluded that no single factor alone can fully explain a game’s performance, emphasizing the need to analyze multiple attributes together.

Building on this prior work, we aim to examine how factors such as Metacritic critic scores, Metacritic user scores, publisher portfolio size, and multi-platform releases relate to and predict total sales for action games released between 2001 and 2020. Metacritic started in 2001 so we will not include any titles released before then. We chose to focus on action games because they consistently dominate global sales revenues and provide a strong basis for analyzing how critical reception, publisher influence, and platform availability contribute to commercial success. Additionally, we seek to determine whether these same relationships can be used to predict total sales for adventure games, allowing us to understand market differences between action and adventure genres.

<a name="ref1"></a><sup>1</sup> Kraemer, T., et al. (2024). Do all stars shine the same? Investigating the nonlinear effects of user and critic reviews on video game sales. Journal of Business Research. https://www.sciencedirect.com/science/article/pii/S0148296324005381#s0075

<a name="ref1"></a><sup>2</sup> Zhang, X., et al. (2021). Evaluation of Game Publishers and Study on the Player Preference Based on the Data from 1980 to 2020 in Video Game Market. Association for Computing Machinery. https://dl.acm.org/doi/fullHtml/10.1145/3463858.3463870
