## Research Question

Do higher critic scores, larger publisher size, and multi-platform releases predict higher total sales among Action games from 2000–2019?


## Background and Prior Work

The video game industry has become one of the largest entertainment industries in the world, with game sales influenced by a variety of factors beyond gameplay itself. Features such as critic scores, publisher resources, and platform accessibility can all affect whether a game achieves commercial success. Among all genres, action games have consistently ranked among the best-selling and most popular from 2000–2019, making them an effective focus for studying what contributes to higher total sales. 

Previous research has shown that critic reviews are closely connected to video game sales. For example, John Sacranie’s study Consumer Perceptions & Video Game Sales: A Meeting of the Minds found that review scores from professional critics played a major role in consumer purchasing decisions and were positively associated with stronger sales performance. Similarly, research by Kaimann and Cox demonstrated that professional critic reviews significantly influenced video game sales, even more strongly than user reviews. <a name="ref1"></a><sup>1</sup> These findings suggest that higher critic scores may help predict higher 
total sales among video games.

Research from a Carnegie Mellon University project also examined how characteristics such as publisher influence, genre, and ratings affect game performance in the market. Their findings suggested that games produced by larger publishers tend to sell more successfully due to stronger marketing power, greater brand recognition, and wider distribution networks. <a name="ref2"></a><sup>2</sup> This prior work connects directly to our investigation of whether publisher size predicts total sales, since larger publishers may have advantages that increase a game’s commercial reach.

In addition, prior studies have highlighted the importance of platform availability in determining market success. Games released on multiple platforms can reach broader audiences by allowing players across different consoles and systems to access the same title. Research on gaming industry trends suggests that wider accessibility often contributes to increased sales opportunities and stronger overall market performance. <a name="ref3"></a><sup>3</sup> 

Building on this prior research, our project investigates whether higher critic scores, larger publisher size, and multi-platform releases predict higher total sales among action games released between 2000 and 2019. We hope to better understand whether critic score, publisher size, and platform reach can serve as meaningful predictors of a game’s market performance.
