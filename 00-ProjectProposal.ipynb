## Research Question

What game attributes (critic score, publisher size, platform) are associated with regional differences in Action game sales across North America, Japan, and PAL territories, using VGChartz sales data from 2000–2019? We might do like a top 100 or narrow down the count of games to make analyzing simpler.

## Background and Prior Work

The video game industry is a global market where player preferences can vary significantly by region. Differences in culture, gaming platforms, and social trends influence the types of games players enjoy, as well as how they engage with features such as storylines, characters, and customization options. Gaming is widespread across the world, with players spanning multiple countries and continents, demonstrating that video games are a shared but culturally influenced activity.<a name="ref1"></a><sup>1</sup> Among all genres, action games consistently rank as one of the most popular worldwide, making them a useful focus for analyzing regional trends and player behavior.

Previous research has explored how game sales and ratings differ across regions. For example, a study by Casey Hoffman at NYU analyzed video game sales data to identify regional preferences for different genres, finding that certain genres perform better in specific regions due to cultural and market differences.<a name="ref2"></a><sup>2</sup> This work highlights how sales data can serve as a proxy for player interest, but it primarily focuses on broad genre categories rather than deeper attributes within a single genre.

Other studies and industry reports have shown that player engagement goes beyond genre alone. Features such as character design, storytelling, and in-game customization (e.g., skins) play a major role in player satisfaction and retention. Additionally, video games are played by a wide range of demographics, with over 200 million players in the United States alone, showing how broadly interactive entertainment reaches across different groups.<a name="ref3"></a><sup>3</sup> These elements are especially relevant in action and shooter games, where personalization and immersion are key aspects of the experience.

Building on this prior work, we're interested in finding the regional preferences for action games and ranking them by their sales of copies sold. We're narrowing it down to just action games because it's the most popular genre and we're curious to see what other attributes besides critics ratings are available to analyze. From our own experiences playing action and shooter games, we generally identify with the player-base of North America, and would occasionally bump into players from EU, Japan, China, Korean, and other regions. We're curious about their preferences about the same games that we play, such as skins, characters, storylines, and more. We hope through our data analysis, to better understand how cultural differences influence what players value in the same games.

## Data

**Ideal Dataset** <br>
To fully answer the research question of what game attributes are associated with regional differences in Action game sales across North America, Japan, and PAL territories, the ideal dataset would be a single, comprehensive, game-level table where every record represents one unique Action game title with no missing values across all relevant variables. <br>
**Variables:** <br>The ideal dataset would contain the following variables. The outcome variables would be regional sales figures, specifically, units sold separately for the major gaming regions like North America, Japan, European Union, Australia, Africa and Asia, as well as each region's share of a title's total global sales. This can allow for meaningful comparison across markets of unequal size. 
The primary explanatory variables would be:
- **Critic score:** A standardized rating between 0-10 or 0-100 from a consistent aggregator such as Metacritic. 
- **Publisher size:** Based on the total number of titles a publisher has released, distinguishing major studios from smaller publishers.
- **Platform or console:** Something liek PlayStation, Xbox, Nintendo DS
Additional control variables would include factors like release year, to account for the overall growth of gaming markets over time; developer country of origin, since a Japanese-developed game may sell differently in Japan regardless of genre; localization status, meaning whether the game was translated and culturally adapted for each region; and launch price per region, to control for affordability differences across markets.

**Observations:**
A minimum of 300–500 complete observations that have unique game titles with no missing values on any key variable would be needed to conduct reliable regression analysis and make subgroup comparisons across regions and platforms. Having at least 500 complete cases would allow for meaningful splits like comparing major vs. minor publishers, or pre- vs. post-2010 releases, without losing statistical power. Ideally, about 1,000 observations would be available so that the results remain robust even after restricting to the 2000-2019 window.

**Collection:**
In the ideal scenario, sales data would be sourced directly from the console manufacturers like Sony, Microsoft, Nintendo and major digital storefronts like Steam, PlayStation Store, Nintendo eShop, which record exact unit sales per region at the transaction level. Critic scores would be pulled programmatically from Metacritic's API. Publisher metadata, developer country of origin, and localization language support would come from IGDB (Internet Game Database), a curated games metadata API. Launch pricing by region would be collected from historical pricing databases or archived storefront records. All data collection would be automated and reproducible via scripts, ensuring that the dataset can be updated as new titles are released.
Storage and organization

**Storage and organization:**
The ideal dataset would be structured as a flat, analysis-ready CSV file with one row per unique game title, where multi-platform releases have been aggregated so that sales figures represent a title's total across all platforms. A secondary relational table would preserve the platform-level breakdown like linking each title to its individual console releases allowing users to reconstruct platform-specific figures if needed. The main table and the platform lookup table would be joinable via a unique title identifier. All monetary and sales figures would be stored in consistent units like millions of copies sold, and all categorical variables, such as publisher and platform, would be consistently coded with no free-text variation.

**Datasets**<br>
- **Dataset 1: Video Game Sales 2024:** This dataset is available at kaggle.com/datasets/asaniczka/video-game-sales-2024 and is free to download with a Kaggle account. The data was scraped from VGChartz.com, a community-maintained database of global video game sales, and covers over 64,000 titles from 1971 to 2024. The variables directly relevant to this project are title, genre, console, publisher, developer, critic_score, na_sales, jp_sales, pal_sales, other_sales, total_sales, and release_date. After filtering to Action titles released between 2000 and 2019 and aggregating multi-platform releases by title, the dataset yields 3,548 unique titles, of which 619 have complete data on both regional sales and critic score, which is the analytic sample for this project. This is the primary dataset for the project and comes closest to the ideal, but it falls short in several ways: developer country of origin, localization status, and launch price are entirely absent; publisher size must be derived rather than directly observed, this can possibly be done by counting how many titles each publisher appears on; and critic score is missing for roughly 83% of Action titles in the window, which is the most significant limitation relative to the ideal dataset.
- **Dataset 2: Metacritic Game Ratings:** A supplementary Metacritic dataset is available on Kaggle, like kaggle.com/datasets/destring/metacritic-reviewed-games-since-2000, and is free to access with a Kaggle account. This dataset provides Metacritic critic scores and user scores for thousands of games released from 2000 onward, including variables such as title, platform, release_date, metascore, and user_score. It does not contain any sales data and therefore cannot stand alone for this project, but it could be joined to the VGChartz dataset on title and release year to fill in the large proportion of missing critic scores in Dataset 1. The key challenge is that title names are not standardized across sources, so some matches might be incorrect.
- **Dataset 3: IGDB (Internet Game Database):** API IGDB is accessible at api-docs.igdb.com and is free to use at the standard tier, though it requires registering a Twitch developer account to obtain OAuth credentials which is relatively easy to do. IGDB is a richly curated database covering developer country of origin, publisher classification, franchise membership, and supported languages, which can be a proxy for localization. These variables address two of the most important gaps in the VGChartz dataset which is developer origin and localization status. The practical limitation is that IGDB data must be retrieved through API calls rather than downloaded as a flat file, meaning a data collection script would need to be written to query each title and store the results. It also does not contain sales figures, so it would function purely as an enrichment source joined to the main dataset by title.
